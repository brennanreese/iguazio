# MLRun Artifacts

Artifacts: data that is produced and/or consumed by functions/jobs/pipelines<br>
Artifact metadata is stored in the mlrun database<br>

Files — files, directories, images, figures, and plotlines.<br>
Datasets — any data , such as tables and DataFrames.<br>
Models — all trained models.<br>
Feature Store Objects - Feature Sets and Feature Vectors

## Artifact path

```python
from os import path
from mlrun import set_environment, get_or_create_project

project_name, artifact_path = set_environment(project="mlrun-artifacts", artifact_path="./artifacts")
project = get_or_create_project(name=project_name, context=path.abspath('./'))
```

```python
# mlrun: start-code
```

```python
def handler(context):
    context.log_artifact('myfile', body="some data", local_path="myfile.txt")
```

```python
# mlrun: end-code
```

```python
from mlrun import code_to_function
fx = code_to_function("create_file", 
                       project="mlrun-artifacts", 
                       filename="mlrun_artifacts.ipynb", 
                       handler="handler", 
                       kind="job",
                       image="mlrun/mlrun")
                       
fx.run(name="create_file", handler="handler", local=True)
```

```python
set_environment(artifact_path="v3io:///projects/{{project}}/artifacts")
fx.run(name="create_file", handler="handler")
```

## Artifact URIs, Metadata, & Versioning

```python
# log_artifact URI = store://<type>/<project>/<key/path>[:tag]
# artifacts are versioned with UIDs, and have tags (use :latest if not specified)
# metadata includes: name, project, updateDate, version, producedBy, sources, formats, etc.
from mlrun import datastore, run
artifact = 'store://datasets/mlrun-functions/prep-func_cleaned_data#0:latest@bf2a6d217fdf4199aa4056e08d4cf33f'
ds = datastore.get_store_resource(artifact)
print(ds.stats)

di = run.get_dataitem(artifact)
di.as_df().head()
```
